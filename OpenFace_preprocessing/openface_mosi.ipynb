# 1.1 生成命令

```python
import os

def list_files_in_directory(directory):
    root_folder = []
    file_paths = []  # Generating commands

    # 遍历指定文件夹中的所有文件和子文件夹
    for root, dirs, files in os.walk(directory):
        root_folder.append(root)
        
        for file in files:
            if not file.startswith('.'):  # 排除以点开头的文件
                file_path = os.path.abspath(os.path.join(root, file))  # 获取文件的绝对路径
                file_paths.append(file_path)  # 将文件的绝对路径添加到列表中

    return root_folder, file_paths

root = r'D:\datasets\MOSI\Raw' # 数据集源数据路径
root_folder_list, _ = list_files_in_directory(root)
root_folder_list = root_folder_list[1:]
# print(root_folder_list)
# print(len(root_folder_list))

# 总共93个文件夹，index一直到92
folder_path = root_folder_list[92]


# print(root_folder)
# # 调用函数并传入要遍历的文件夹路径
# folder_path = r'D:\datasets\SIMS\Raw\video_0060'  # 修改为你的文件夹路径
_, files = list_files_in_directory(folder_path)

commandlist = []
# 输出所有文件的绝对路径
for file_path in files:
    commandlist.append('-f '+ '\''+file_path+'\'')

command = '.\FeatureExtraction.exe'
for command_dir in commandlist:
    command = command + ' ' + command_dir


print(command)


```

# 1.2 删去多余的avi和hog文件，并重命名

```python
def delete_avi_files_and_rename(directory, folder_name):
    # 遍历指定文件夹中的所有文件和子文件夹
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.avi'):  # 检查文件是否以.avi结尾
                file_path = os.path.join(root, file)  # 获取文件的路径
                os.remove(file_path)  # 删除文件
            if file.endswith('.hog'):  # 检查文件是否以.avi结尾
                file_path = os.path.join(root, file)  # 获取文件的路径
                os.remove(file_path)  # 删除文件
    # 重命名文件夹为"audio"
    os.rename(directory, os.path.join(os.path.dirname(directory), folder_name))
    
# 调用函数并传入要删除的文件夹路径
gen_folder_path = r'D:\OpenFace_2.2.0_win_x64\OpenFace_2.2.0_win_x64\processed'  # 修改为你的文件夹路径
folder_name = folder_path.split('\\')[-1]
print(folder_name)
delete_avi_files_and_rename(gen_folder_path, folder_name)
```

```python
import pandas as pd

AU_to_des = {' AU01_c': 'raise inner brow', 
             ' AU02_c':'raise outer brow', 
             ' AU04_c':'lower brow', 
             ' AU05_c':'raise upper lid', 
             ' AU06_c':'raise cheek', 
             ' AU07_c':'tighten lid', 
             ' AU09_c':'wrinkle nose', 
             ' AU10_c':'raise upper lip', 
             ' AU12_c':'pull lip corner', 
             ' AU15_c':'depress lip corner', 
             ' AU20_c':'stretch lip', 
             ' AU23_c':'tighten lip', 
             ' AU25_c':'part lip', 
             ' AU26_c':'drop jaw', 
             ' AU28_c':'suck lip', 
             ' AU45_c':'blink'}

# AU_to_des = {' AU01_c': '抬起内眉', 
#              ' AU02_c':'抬起外眉', 
#              ' AU04_c':'低眉', 
#              ' AU05_c':'抬起上眼睑', 
#              ' AU06_c':'抬起脸颊', 
#              ' AU07_c':'紧闭眼皮', 
#              ' AU09_c':'皱鼻头', 
#              ' AU10_c':'扬起上唇', 
#              ' AU12_c':'拉唇角', 
#              ' AU15_c':'下压唇角', 
#              ' AU20_c':'拉伸嘴唇', 
#              ' AU23_c':'抿紧嘴唇', 
#              ' AU25_c':'分开嘴唇', 
#              ' AU26_c':'掉下巴', 
#              ' AU28_c':'嘬嘴', 
#              ' AU45_c':'眨眼'}

# -----------函数find_AU:找出csv文件中符合要求的AU------------
def find_AU(path):

    AUlist = []
    
    # 读取 CSV 文件
    all_data = pd.read_csv(path)

    # 获取文件第一行（列名）
    column_names = all_data.columns.tolist()

    # 找到目标列的索引
    target_columns = [' AU01_c', ' AU02_c', ' AU04_c', ' AU05_c', ' AU06_c', ' AU07_c', ' AU09_c', ' AU10_c', ' AU12_c'
                     , ' AU15_c', ' AU20_c', ' AU23_c', ' AU25_c', ' AU26_c', ' AU28_c', ' AU45_c']
    selected_columns = [idx for idx, col in enumerate(column_names) if col in target_columns]

    # 提取目标列的数据
    data = all_data.iloc[:, selected_columns]

    # # 显示提取的数据
    # print(data)
    # print(type(data))


    # 找到符合要求的列
    selected_columns = []

    for column in data.columns:
        consecutive_ones = 0
        max_consecutive_ones = 3  # 至少连续出现3个1（选取持续超过3帧的表情）
        found = False

        for value in data[column]:
            if value == 1:
                consecutive_ones += 1
                if consecutive_ones >= max_consecutive_ones:
                    found = True
                    break
            else:
                consecutive_ones = 0

        if found:
            selected_columns.append(column)

    # 输出符合要求的列名
#     print("符合要求的列名：", selected_columns)

    # 提取符合要求的列的数据
    selected_data = data[selected_columns]

    # 计算每列的数值和
    sums = selected_data.sum()

    # 建立字典，key为列名，value为数值和
    sums_dict = sums.to_dict()

    # 按照值从大到小排名
    sorted_sums = sorted(sums_dict.items(), key=lambda x: x[1], reverse=True)

    # 输出排序后的结果
    for col, total in sorted_sums:
        print(f"列名: {col}, 数值和: {total}")

    # 或者将结果存储到新的字典中
    sorted_sums_dict = dict(sorted_sums)
    sorted_sums_dict = dict(list(sorted_sums_dict.items())[:8]) # 选取前k个描述
    # 显示提取的数据
    return sorted_sums_dict

# 根据路径生成id
def generate_id(path):
    # 从文件路径中提取视频和文件名
    parts = path.split("\\")

    video_name = ""
    file_name = ""

    video_name = parts[-2]
    for part in parts:
        if part.endswith(".csv"):
            file_name = part[:-4]  # 去除文件扩展名 .csv
    print(file_name)
    # 创建 ID
    if video_name and file_name:
        generated_id = f"{video_name}$_${file_name}"
    else:
        print("未能生成 ID，文件路径格式不匹配。")
    return generated_id



vision_text_list = []
for path in csv_files:
    print(path)
    selected_data = find_AU(path)
    
    description = []
    for au in selected_data.keys():
        if au in AU_to_des.keys():
            description.append(AU_to_des[au])

#     print(description)
#     vision_text = 'The speaker made such an expression: '
    vision_text = 'The speaker made such an expression: '
    for des in description:
        vision_text = vision_text + des + ', ' 
    
    # 制作字典{'id':xxx, 'raw_vision_text':xxx}
    raw_id = generate_id(path)
    dic = {'id':raw_id, 'raw_vision_text':vision_text}
    print(dic)
    vision_text_list.append(dic)
    
print(len(vision_text_list))
```

# # 计算MOSI的vision_text平均长度

```python
import pandas as pd
import numpy as np
import csv
import pickle


def save_pkl(path, obj):
    pickle_file = open(path, 'wb')
    pickle.dump(obj, pickle_file)
    pickle_file.close()
    print("保存成功")

def load_pkl(path):
    pickle_file = open(path, 'rb')
    obj = pickle.load(pickle_file)
    pickle_file.close()
    print("读取成功")
    return obj

data = load_pkl(r'D:\OpenFace_2.2.0_win_x64\OpenFace_2.2.0_win_x64\MOSI\vision_text_en_max8.pkl')
print(data[0:10])
max_len = 0
length = 0
for item in data:
    if len(item['raw_vision_text'].split()) > max_len:
        max_len = len(item['raw_vision_text'].split())
#     print(len(item['raw_vision_text']))
    length += len(item['raw_vision_text'].split())

mean_len = length / len(data)
print(mean_len)
print('max len:{}'.format(max_len))
```
