```python
def delete_avi_files_and_rename(directory):
    # Iterates through all files and subfolders in the specified folder
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.avi'):  # Check if the file ends with .avi
                file_path = os.path.join(root, file)  # Get the path to the file
                os.remove(file_path)  
            if file.endswith('.hog'):  # Check if the file ends with .avi
                file_path = os.path.join(root, file)  # Get the path to the file
                os.remove(file_path)  
    # Rename folder to "audio"
    os.rename(directory, os.path.join(os.path.dirname(directory), 'video_0060'))
    
# Call the function and pass in the path to the folder to delete
folder_path = r'D:\OpenFace_2.2.0_win_x64\OpenFace_2.2.0_win_x64\processed'  # Change it to your folder path.
delete_avi_files_and_rename(folder_path)
```

```python
import os

def get_folders_in_directory(directory):
    folder_names = []  # List for storing folder names

    # Iterates through all files and folders in the specified folder
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if os.path.isdir(item_path):  # Check if it is a folder.
            folder_names.append(item)  # Add the folder name to the list

    return folder_names

# Call the function and pass in the folder path to get the folder name
folder_path = r'D:\datasets\SIMS\Raw'  # Change it to your folder path.
folder_list = get_folders_in_directory(folder_path)

# Output folder name list
print(folder_list)
```

# 2.1 找出csv文件中符合要求的AU

```python
import pandas as pd

# 函数find_AU:找出csv文件中符合要求的AU
def find_AU(path):

    AUlist = []
    
    # 读取 CSV 文件
    all_data = pd.read_csv(path)

    # 获取文件第一行（列名）
    column_names = all_data.columns.tolist()

    # 找到目标列的索引
    target_columns = [' AU01_c', ' AU02_c', ' AU04_c', ' AU05_c', ' AU06_c', ' AU07_c', ' AU09_c', ' AU10_c', ' AU12_c'
                     , ' AU15_c', ' AU20_c', ' AU23_c', ' AU25_c', ' AU26_c', ' AU28_c', ' AU45_c']
    selected_columns = [idx for idx, col in enumerate(column_names) if col in target_columns]

    # 提取目标列的数据
    data = all_data.iloc[:, selected_columns]

    # # 显示提取的数据
    # print(data)
    # print(type(data))


    # 找到符合要求的列
    selected_columns = []

    for column in data.columns:
        consecutive_ones = 0
        max_consecutive_ones = 3  # 至少连续出现3个1
        found = False

        for value in data[column]:
            if value == 1:
                consecutive_ones += 1
                if consecutive_ones >= max_consecutive_ones:
                    found = True
                    break
            else:
                consecutive_ones = 0

        if found:
            selected_columns.append(column)

    # 输出符合要求的列名
    print("符合要求的列名：", selected_columns)

    # 提取符合要求的列的数据
    selected_data = data[selected_columns]

    # 显示提取的数据
    return selected_data

path = r'D:\OpenFace_2.2.0_win_x64\OpenFace_2.2.0_win_x64\SIMS\video_0053\0053.csv'
selected_data = find_AU(path)
print(selected_data)


# 计算每列的数值和
sums = selected_data.sum()

# 建立字典，key为列名，value为数值和
sums_dict = sums.to_dict()

# 按照值从大到小排名
sorted_sums = sorted(sums_dict.items(), key=lambda x: x[1], reverse=True)

# 输出排序后的结果
for col, total in sorted_sums:
    print(f"列名: {col}, 数值和: {total}")

# 或者将结果存储到新的字典中
sorted_sums_dict = dict(sorted_sums)
print(sorted_sums_dict)
```
