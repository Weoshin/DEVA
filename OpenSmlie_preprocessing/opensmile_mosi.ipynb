```python
import os
from moviepy.editor import VideoFileClip

def extract_audio(video_path, audio_path):
    video = VideoFileClip(video_path)
    audio = video.audio
    audio.write_audiofile(audio_path)
    
# 输出：mp4的路径
def list_files_in_directory(directory):
    file_paths = []  # 用于存储文件路径的列表

    # 遍历指定文件夹中的所有文件和子文件夹
    for root, dirs, files in os.walk(directory):
        for file in files:
            if not file.startswith('.'):  # 排除以点开头的文件
                file_path = os.path.abspath(os.path.join(root, file))  # 获取文件的绝对路径
                file_paths.append(file_path)  # 将文件的绝对路径添加到列表中

    return file_paths

def get_folders_in_directory(directory):
    folder_names = []  # 用于存储文件夹名的列表

    # 遍历指定文件夹中的所有文件和文件夹
    for item in os.listdir(directory):
        item_path = os.path.join(directory, item)
        if os.path.isdir(item_path):  # 检查是否为文件夹
            folder_names.append(item)  # 将文件夹名添加到列表中

    return folder_names

# 调用函数并传入要获取文件夹名的文件夹路径
folder_path = r'D:\datasets\MOSI\Raw'  # 修改为你的文件夹路径
folder_list = get_folders_in_directory(folder_path)

# 输出文件夹名列表
print(folder_list)

for folder in folder_list:
    folder_path = r'D:\datasets\MOSI\Raw\\' + folder  # 修改为你的文件夹路径\
    files = list_files_in_directory(folder_path)
    os.makedirs(r'D:\datasets\MOSI\Raw_wav\\' + folder)
    for file in files:
#         print(file)
        mp4name = file.split('\\')[-1]
        wavname = mp4name.split('.')[0] + '.wav'
        wavsave_path = r'D:\datasets\MOSI\Raw_wav\\' + folder + '\\' + wavname
#         print('input path:{}'.format(file))
#         print('output path:{}'.format(wavsave_path))
        extract_audio(file, wavsave_path)
    print('folder:{} complete'.format(folder))
```

```python
'''
loudness:响度
jitter：频率变化率
shimmer：振幅变化率
f0final：基频均值
'''
import os
import pandas as pd

def find_key_value(path):
    data = pd.read_csv(path, sep='\t', header=0)
    value = data.iloc[-1,:]

    split_data = [num for num in value[0].split(",")]  
    split_value = []
    for item in split_data[1:-1]:  
        split_value.append(float(item))

#     print('length:{}'.format(len(split_value))) # 共有1582条数据，对应着key的行数从4到1585
#     for item in split_value:  
#         print(item)
    
    target_name = ["@attribute pcm_loudness_sma_amean numeric" ,"@attribute jitterLocal_sma_amean numeric",
              "@attribute shimmerLocal_sma_amean numeric", "@attribute F0final_sma_amean numeric"]
    for name in target_name:
        row_index = data[data["@relation openSMILE_features"] == name].index.tolist()[0]
#         print(row_index)
        
    key_value = {"loudness_sma_amean": split_value[2],"jitterLocal_sma_amean": split_value[1449],
              "shimmerLocal_sma_amean": split_value[1487], "F0final_sma_amean": split_value[1430]}
#     print(key_value)
    
    wav_id = path.split('\\')[-2] + '$_$' + path.split('\\')[-1][:-4]
#     print(wav_id)
    LJSF_dic = {wav_id: key_value}
    return LJSF_dic


root_path = r'D:\opensmile-3.0-win-x64\opensmile-3.0-win-x64\MOSI\\'
root_list = os.listdir(root_path)
print(root_list)

LJSF = dict()
for root in root_list:
    audio_path = os.path.join(root_path, root) # D:\opensmile-3.0-win-x64\opensmile-3.0-win-x64\SIMS\\video_0001
    audio_list = os.listdir(audio_path)
    for audio in audio_list:
        input_path = os.path.join(audio_path, audio)
        print(input_path)
        LJSF_dic = find_key_value(input_path)
        LJSF = {**LJSF, **LJSF_dic}

print(LJSF)
#     print(audio_path)
#     LJSF_dic = find_key_value(r'D:\opensmile-3.0-win-x64\opensmile-3.0-win-x64\SIMS\video_0001\0002.csv')
#     print(LJSF_dic)
```

# 根据均值确定低中高，生成自然语言描述

```python
jitter_list = []
loudness_list = []
shimmer_list = []
f0_list = []

for key, value in LJSF.items():
    
    loudness_list.append(value['loudness_sma_amean'])
    jitter_list.append(value['jitterLocal_sma_amean'])
    shimmer_list.append(value['shimmerLocal_sma_amean'])
    f0_list.append(value['F0final_sma_amean'])
    
print(loudness_list[0])
loudness_list = sorted(loudness_list)
jitter_list = sorted(jitter_list)
shimmer_list = sorted(shimmer_list)
f0_list = sorted(f0_list)

mid_point = len(loudness_list) // 3                 # 760
upper_limit = mid_point + len(loudness_list) // 3   # 1520
# loudness: 
print('loudness_min:{}'.format(loudness_list[0]))
print('loudness_mid:{}'.format(loudness_list[mid_point]))
print('loudness_upper:{}'.format(loudness_list[upper_limit]))
print('loudness_max:{}'.format(loudness_list[-1]))
Lm = loudness_list[mid_point]
Lu = loudness_list[upper_limit]
# jitter:
print('jitter_min:{}'.format(jitter_list[0]))
print('jitter_mid:{}'.format(jitter_list[mid_point]))
print('jitter_upper:{}'.format(jitter_list[upper_limit]))
print('jitter_max:{}'.format(jitter_list[-1]))
Jm = jitter_list[mid_point]
Ju = jitter_list[upper_limit]
# shimmer:
print('shimmer_min:{}'.format(shimmer_list[0]))
print('shimmer_mid:{}'.format(shimmer_list[mid_point]))
print('shimmer_upper:{}'.format(shimmer_list[upper_limit]))
print('shimmer_max:{}'.format(shimmer_list[-1]))
Sm = shimmer_list[mid_point]
Su = shimmer_list[upper_limit]
# f0:
print('f0_min:{}'.format(f0_list[0]))
print('f0_mid:{}'.format(f0_list[mid_point]))
print('f0_upper:{}'.format(f0_list[upper_limit]))
print('f0_max:{}'.format(f0_list[-1]))
Fm = f0_list[mid_point]
Fu = f0_list[upper_limit]
```

```python
def gene_loudness(loudness, Lm, Lu):
    if loudness < Lm:
        Lat = 'low loudness'
    elif loudness >= Lm and loudness < Lu:
        Lat = 'normal loudness'
    elif loudness >= Lu:
        Lat = 'high loudness'
    return Lat

def gene_jitter(jitter, Jm, Ju):
    if jitter < Jm:
        Jat = 'low jitter'
    elif jitter >= Jm and jitter < Ju:
        Jat = 'normal jitter'
    elif jitter >= Ju:
        Jat = 'high jitter'
    return Jat

def gene_shimmer(shimmer, Sm, Su):
    if shimmer < Sm:
        Sat = 'low shimmer'
    elif shimmer >= Sm and shimmer < Su:
        Sat = 'normal shimmer'
    elif shimmer >= Su:
        Sat = 'high shimmer'
    return Sat

def gene_f0(f0, Fm, Fu):
    if f0 < Fm:
        Fat = 'low pitch'
    elif f0 >= Fm and f0 < Fu:
        Fat = 'normal pitch'
    elif f0 >= Fu:
        Fat = 'high pitch'
    return Fat

audio_text_list = []
for key, value in LJSF.items():
    Lat = gene_loudness(value['loudness_sma_amean'], Lm, Lu)
    Jat = gene_jitter(value['jitterLocal_sma_amean'], Jm, Ju)
    Sat = gene_shimmer(value['shimmerLocal_sma_amean'], Sm, Su)
    Fat = gene_f0(value['F0final_sma_amean'], Fm, Fu)
    audio_text = 'The speaker speaks in the following tone: ' + Lat + ', ' + Jat + ', ' + Sat + ', ' + Fat + '.'
    dic = {'id': key, 'raw_audio_text': audio_text}
    audio_text_list.append(dic)


print(len(audio_text_list))
print(audio_text_list[0])

```
